```python
import pandas as pd 
```

```python
students_df = pd.read_csv('student_records_with_total_marks.csv')
print("students_df:")
students_df.head()
# Read the CSV file
```

```python
students_df['total_marks'] = students_df[
    ['Sub1 (Marks)', 'Sub2 (Marks)', 'Sub3 (Marks)', 'Sub4 (Marks)', 'Sub5 (Marks)', 'Sub6 (Marks)']
].sum(axis=1)

students_df.head()
# Total marks by The branches ABOVE 5 student (So I'm using head function)
```

```python
cs_students = students_df[students_df['Branch'] == 'CS']
cs_students
# Print student who is only from the CSE branch.( AND TOTAL MARKS)
```

```python
cs_students = students_df[students_df['Branch'] == 'DS']
cs_students
# Define and print which student Which is belonging to 2 CSE(DS)
```

```python
cse_ds_students = students_df[
    (students_df['Branch'] == 'CS') &
    (students_df['Dept.'] == 'DS')
]
cse_ds_students
```

```python
years_22to_25 = students_df[students_df['Year'].isin([2021,2022, 2023, 2024, 2025])]
years_22to_25
# Get student who is studying from 2021 to 2025
```

```python
years_cs_21to25 = students_df[
    (students_df['Year'].isin([2021, 2022, 2023, 2024, 2025])) &
    (students_df['Branch'] == 'CS')
]

years_cs_21to25
```

```python
years_cs_21to25 = students_df[
    (students_df['Year'].isin([2021, 2022, 2023, 2024, 2025])) &
    (students_df['Branch'] == 'DS')
]

years_cs_21to25
```

```python
students_lucknow_gaziabad = students_df[
    students_df['City'].isin(['Lucknow', 'Ghaziabad','Gorakhpur'])
]
students_lucknow_gaziabad
```

```python
top_students_per_year = students_df.loc[
    students_df.groupby('Year')['total_marks'].idxmax()
]
top_students_per_year
```

```python
year_branch_stats = students_df.groupby(['Year', 'Branch'])['total_marks'].agg(['sum', 'mean']).reset_index()

print(" Yearly sum & average total marks per branch")
year_branch_stats
```

```python
max_total_marks = 600  # 6 subjects  100 marks each

students_df['percentage'] = (students_df['total_marks'] / max_total_marks) * 100

print(" Students with calculated percentage")
students_df.head()
```

```python
low_scorers = students_df[
    (students_df[['Sub1 (Marks)','Sub2 (Marks)','Sub3 (Marks)','Sub4 (Marks)','Sub5 (Marks)','Sub6 (Marks)']] < 50).any(axis=1)
]
low_scorers
```

```python
# Select only relevant columns to display clearly:
low_scorers_clean = low_scorers[[
    'Roll No.', 'Branch', 'Year',
    'Sub1 (Marks)','Sub2 (Marks)','Sub3 (Marks)','Sub4 (Marks)','Sub5 (Marks)','Sub6 (Marks)'
]]
```

```python
city_stats = students_df.groupby('City')['total_marks'].mean().reset_index()
top_city = city_stats.loc[city_stats['total_marks'].idxmax()]

print(" City with highest average student marks", top_city)
city_stats
```

```python
cs_students_by_year = students_df[students_df['Branch'] == 'CS'].groupby('Year').size()
print("Number of CS students per year", cs_students_by_year)
```

```python
cs_students = students_df[students_df['Branch'] == 'CS']

top_cs_students_per_year = cs_students.loc[
    cs_students.groupby('Year')['total_marks'].idxmax()
]

top_cs_students_per_year
```

```python
city_avg = students_df.groupby('City')['total_marks'].mean().reset_index()

top_city = city_avg.loc[city_avg['total_marks'].idxmax()]

print(" City with highest average marks:", top_city)
city_avg
```

```python
import matplotlib.pyplot as plt
city_avg.plot.barh(x='City', y='total_marks', legend=False, figsize=(6,4), color='salmon', title='Avg Marks by City')
```
